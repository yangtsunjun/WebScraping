# 抓取資料

```python
# retreive ptt movieweb page's source code
import urllib.request as req

url = 'https://www.ptt.cc/bbs/movie/index.html' # url of ptt movie web page

# 建立一個 Request 物件，附加 request headers 的資訊!!!!
request = req.Request(url, headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
})

# 利用 request物件 而不是url 去打開網址
with req.urlopen(request) as response:
    data = response.read().decode('utf-8')
print(data)
```

## 解析原始碼資料

```python
# use beautifulsoup4 to capture page data
import bs4
root = bs4.BeautifulSoup(data, 'html.parser') # let BeautifulSoup to 解析 HTML files 

titles = root.find_all('div', class_='title') # 尋找所有 class='title' 的 div 標籤

for title in titles:
    if title.a != None: # print title if 開頭不是 a 
        print(title.a.string) # a代表非刪除的文章，string就是其中的文字
        
```
